```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
df=pd.read_csv('example_1.csv')
df.head(20)
```

```
time=df[df.columns[0]]
t=[]
for j in range(0,len(time)):
    t.append(j)

x=df[df.columns[1]]
y=df[df.columns[2]]
z=df[df.columns[3]]
plt.subplot(3, 1, 1)
plt.xticks(t, " ")
plt.plot(t, x)
plt.title('Standard')
plt.ylabel('PM 1')

plt.subplot(3, 1, 2)
plt.xticks(t, " ")
plt.plot(t, y,color='red')
plt.xlabel('time (s)')
plt.ylabel('PM 2.5')

plt.subplot(3, 1, 3)
plt.xticks(t, " ")
plt.plot(t, z,color='green')
plt.xlabel('time (s)')
plt.ylabel('PM 10')
plt.figure(num=None, figsize=(20, 10), dpi=80, facecolor='w', edgecolor='k')
plt.show()
```
